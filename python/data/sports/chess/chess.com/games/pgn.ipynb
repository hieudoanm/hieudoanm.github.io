```python
import chess.pgn
import io
import os

folders = [f.path for f in os.scandir(".") if f.is_dir()]

for folder in folders:
    files = [f.path for f in os.scandir(folder) if f.is_file()]
    for file in files:
        pgn_file = open(file, encoding="utf-8")
        pgn_string = pgn_file.read()
        games = pgn_string.split("\n\n\n")
        for game in games:
            game_info = chess.pgn.read_game(io.StringIO(game))
```
