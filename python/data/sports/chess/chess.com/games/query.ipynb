```python
import json
from pathlib import Path
import requests
import threading
import urllib.request

usernames: list[str] = [
    "dominguezonyoutube",  # Leinier Domínguez
    "fabianocaruana",  # Fabiano Caruana
    "firouzja2003",  # Alireza Firouzja
    "gmwso",  # Wesley So
    "grischuk",  # Alexander Grischuk
    "hikaru",  # Hikaru Nakamura
    "lachesisq",  # Ian Nepomniachtchi
    "levonaronian",  # Levon Aronian
    "liemle",  # Lê Quang Liêm
    "lyonbeast",  # MVL - Maxime Vachier-Lagrave
    "magnuscarlsen",  # Magnus Carlsen
    "sergeykarjakin",  # Sergey Karjakin
    "wonderfultime",  # Le Tuan Minh
]
user_agent = "ArithmeticErrorMozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:127.0) Gecko/20100101 Firefox/127.0"


def get_pgn_by_username(username : str):
    print(username + "\n")
    Path(username).mkdir(parents=True, exist_ok=True)
    archives_url = f"https://api.chess.com/pub/player/{username}/games/archives"
    archives_response: requests.Response = requests.get(
        archives_url, headers={"User-Agent": user_agent}
    )
    archives_data : dict = archives_response.json()
    archives : list[str] = archives_data.get("archives", [])
    archives.reverse()
    times = list(map(lambda archive: archive.split("/")[-2:], archives))
    for time in times:
        [year, month] = time
        # json
        games_json_url = f"https://api.chess.com/pub/player/{username}/games/{year}/{month}"
        games_json_response = requests.get(games_json_url, headers={"User-Agent": user_agent})
        games_json_data = games_json_response.json()
        games: list = games_json_data.get("games", [])
        games_json_file = open(f"{username}/json/{year}-{month}.json", "w")
        games_json_file.write(json.dumps(games, indent=2))
        # pgn
        pgn_url = f"{games_json_url}/pgn"
        filename: str = f"{username}/pgn/{year}-{month}.pgn"
        urllib.request.urlretrieve(pgn_url, filename)


t1 = threading.Thread(target=get_pgn_by_username, args=(usernames[0],))
t2 = threading.Thread(target=get_pgn_by_username, args=(usernames[1],))
t3 = threading.Thread(target=get_pgn_by_username, args=(usernames[2],))
t4 = threading.Thread(target=get_pgn_by_username, args=(usernames[3],))
t5 = threading.Thread(target=get_pgn_by_username, args=(usernames[4],))
t6 = threading.Thread(target=get_pgn_by_username, args=(usernames[5],))
t7 = threading.Thread(target=get_pgn_by_username, args=(usernames[6],))
t8 = threading.Thread(target=get_pgn_by_username, args=(usernames[7],))
t9 = threading.Thread(target=get_pgn_by_username, args=(usernames[8],))
t10 = threading.Thread(target=get_pgn_by_username, args=(usernames[9],))
t11 = threading.Thread(target=get_pgn_by_username, args=(usernames[10],))
t12 = threading.Thread(target=get_pgn_by_username, args=(usernames[11],))

t1.start()
t2.start()
t3.start()
t4.start()
t5.start()
t6.start()
t7.start()
t8.start()
t9.start()
t10.start()
t11.start()
t12.start()

t1.join()
t2.join()
t3.join()
t4.join()
t5.join()
t6.join()
t7.join()
t8.join()
t9.join()
t10.join()
t11.join()
t12.join()

print("Done!")
```
