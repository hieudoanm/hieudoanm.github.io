```python
import json
import requests


response = requests.get("https://api.github.com/emojis")
emojis = response.json()


# Writing JSON data to a file
with open("emojis.json", "w") as json_file:
    json.dump(emojis, json_file, indent=2)


keys: list[str] = list(emojis.keys())
with open("emojis.txt", "w") as file:
    file.write("\n".join(keys))  # Output: apple, banana, cherry
```

```python
import json
import requests
from tqdm.notebook import tqdm

# Open and read JSON file
with open("emojis.json", "r") as file:
    emojis: dict[str] = json.load(file)

items = emojis.items()
print(len(items))

for key, value in tqdm(items):
    response = requests.get(value)  # Send a GET request
    if response.status_code == 200:  # Check if request was successful
        with open(f"./png/{key}.png", "wb") as file:
            file.write(response.content)  # Write the image content to a file
        print(f"Image downloaded as {f"{key}.png"}")
    else:
        print("Failed to download the image")
```
