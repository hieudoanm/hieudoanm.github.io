```python
import pandas
```

```python
raw_data_frame = pandas.read_csv("./raw.csv")
raw_data_frame["country"] = raw_data_frame["country"].apply(lambda x: x.strip())
raw_data_frame[["month_date", "month", "date"]] = raw_data_frame["full_date"].str.split(
    " ", expand=True
)
raw_data_frame["year_month_date"] = (
    raw_data_frame["year"] + "-" + raw_data_frame["month_date"]
)
raw_data_frame = raw_data_frame.drop(columns=["full_date", "month_date"], axis=1)
raw_data_frame = raw_data_frame[
    ["country", "event", "name", "year_month_date", "year", "month", "date"]
]
sorted_raw_data_frame = raw_data_frame.sort_values(by=["year_month_date"])
sorted_raw_data_frame.to_csv("./data.csv", index=False)
sorted_raw_data_frame
```
