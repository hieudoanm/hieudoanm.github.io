```python
import pandas
import requests
```

```python
response = requests.get(
    "https://www.fifa.com/api/ranking-overview?dateId=id14079",
    headers={"User-Agent": "PostmanRuntime/7.32.3"},
)
data: list[dict] = response.json()
rankings = data.get("rankings", [])
rankings
```

```python
items = []

for ranking in rankings:
    rankingItem = ranking.get("rankingItem", {})
    rank = rankingItem.get("rank")
    name = rankingItem.get("name")
    countryCode = rankingItem.get("countryCode")
    items.append({"rank": rank, "name": name, "countryCode": countryCode})
```

```python
rankings_data_frame = pandas.DataFrame(items)
rankings_data_frame.to_csv("./rankings.csv", index=False, header=False)
rankings_data_frame
```
