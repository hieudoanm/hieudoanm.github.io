```python
import pandas
```

```python
file_name = "./datasets.csv"
```

```python
standings_data_frame = pandas.read_csv(file_name)
standings_data_frame = standings_data_frame.sort_values(
    by=["country", "competition_tier", "competition", "season"]
)
standings_data_frame.to_csv(file_name, index=False)
standings_data_frame
```

```python
competitions = standings_data_frame["competition"].unique()
competitions
```

## Premier League

```python
premier_league_data_frame = standings_data_frame[
    standings_data_frame["competition"] == "Premier League"
]
premier_league_data_frame
```

```python
champions = premier_league_data_frame["champion"].value_counts()
champions
```

```python
runners_up = premier_league_data_frame["runner_up"].value_counts()
runners_up
```

```python
third_places = premier_league_data_frame["third_place"].value_counts()
third_places
```
